```python
# Logistic Regression




## The DataSet | Social Network 




## Step 1 | Data Pre-Processing

### Importing the Libraries


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

### Importing the dataset

print("import")


dataset = pd.read_csv(r'C:\Users\PythonE\Desktop\PythonTest\SocialNetworkAds.csv')
#print(dataset)


X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values

print(X)
#print(y)


### Splitting the dataset into the Training set and Test set


from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)


### Feature Scaling
print("feature scaling")

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
print("Xtrain sc fit transorm")
X_test = sc.transform(X_test)
print("just after sc transofmr x test")

print(X_test)


## Step 2 | Logistic Regression Model



### Fitting Logistic Regression to the Training set


from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

classifier.score(X,y)

print("coefficient: \n", classifier.coef_)
print("intercept: \n", classifier.intercept_)
testx=X_test*classifier.coef_ 
print(testx)
testy=testx[:,0] + testx[:,1]+classifier.intercept_
print(testy)

print("odds")
import math
count=1
for x in testy:
    odds=math.exp(x)
    print(count, "prob=", odds/(1+odds))
    count=count+1



#odds= p/ (1-p) = probability of event occurrence / probability of not event occurrence
#ln(odds) = ln(p/(1-p))
#logit(p) = ln(p/(1-p)) = b0+b1X1+b2X2+b3X3....+bkXk


print("fitting")

## Step 3 | Predection

### Predicting the Test set results


y_pred = classifier.predict(X_test)

print(y_pred)


## Step 4 | Evaluating The Predection



### Making the Confusion Matrix


from sklearn.metrics import confusion_matrix
print("import confusion matrix")
cm = confusion_matrix(y_test, y_pred)
print("after getting conf matrix")

print(cm)


## Visualization


```
