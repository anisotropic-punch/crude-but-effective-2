```python
import csv
from sklearn import svm, metrics, datasets
from numpy import genfromtxt
import numpy as np

dataset = genfromtxt(r'C:\Users\PythonE\Desktop\PythonTest\smalltrainrand.csv', delimiter=",", skip_header=1)
labels = [x[0] for x in dataset]
data = [x[1:] for x in dataset]
#for d in data:
    #print(d)
##for d in dataset:
    #print(d)
#print(dataset)
#print(labels)
#print(data)
#break

n_samples = len(labels)
n_features = len(data[0])

print("Number of samples: " + str(n_samples) + ", number of features: "+ str(n_features))

# a support vector classifier
classifier = svm.LinearSVC()

split_point = int(n_samples * 0.66)

# using two thirds for training
# ans one third for testing

labels_learn = labels[:split_point]
data_learn = data[:split_point]

labels_test = labels[split_point:]
data_test = data[split_point:]

print("Training: " + str(len(labels_learn)) + " Test: " + str(len(labels_test)))

# Learning Phase
classifier.fit(data_learn, labels_learn)

# Predict Test Set
predicted = classifier.predict(data_test)
print("label test", labels_test)
print("label pred", predicted)

# classification report
print("Classification report for classifier %s:n%sn" % (classifier, metrics.classification_report(labels_test, predicted)))

# confusion matrix
print("Confusion matrix:n%s" % metrics.confusion_matrix(labels_test, predicted))


len_label=len(labels_test)
for k in range(len_label):
        if labels_test[k] != predicted[k]:
            print(k,"  ",labels_test[k], "    ", predicted[k])
        #print(part[1])
            matr=data_test[k]
            matrp=np.asarray(matr).reshape(28,28)
        #print(matrp)
            c=0
            for x in matrp:
                c=[]
            #print("BBBBBB    ", c)
            #print(x)
                d=x
                for y in d:
                        #print(x[y])
                        #print("cell  ", y)
                    if y > 0:
                        c=c+['X']
                    else:
                        c=c+["."]
                print(c)
     #$$$$$$$$$$ 

print("END")

digits = datasets.load_digits()
print("len=", len(digits))
for d in digits:
    print(digits.images)
```
